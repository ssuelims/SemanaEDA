# Agrupamento (groupby())
# Como agrupar o filme_id calculando a media da avaliacao

```python
df.head()
```

```python
medias_por_filme = df.groupby('filme_id').mean()['avaliacao']
```

```python
medias_por_filme.head()
```

```python
medias_por_filme.plot(kind='hist')
```

```python
medias_por_filme.describe()
```

```python
sns.boxplot(medias_por_filme)
```

# Histplot

```python
sns.histplot(medias_por_filme)
```

# kd = True KDE (linha de densidade kernel) - estimativa da funcao
# bins = 10 - Quantos intervalos eu preciso?

```python
sns.histplot(medias_por_filme,kde=True,bins=10)
```
